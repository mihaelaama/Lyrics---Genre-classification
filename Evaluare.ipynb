```python
import pandas as pd
import joblib
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# Încarcă modelul LDA și vectorizatorul
lda = joblib.load('lda_model.pkl')
vectorizer = joblib.load('vectorizer.pkl')

# Încarcă distribuția topicurilor pe genuri
topic_genre_distribution = pd.read_csv('topic_genre_distribution_for_predict.csv', index_col=0)

# Funcția pentru a prelucra și clasifica noile versuri
def classify_lyrics(lyrics):
    lyrics = " ".join(lyrics.split("\n")).strip()
    X_new = vectorizer.transform([lyrics])
    topic_distribution = lda.transform(X_new)
    dominant_topic = topic_distribution.argmax(axis=1)[0]
    genre = topic_genre_distribution.idxmax(axis=0)[dominant_topic]
    return genre

# Funcția pentru a clasifica un set de versuri
def classify_lyrics_batch(lyrics_list):
    genres = []
    for lyrics in lyrics_list:
        genre = classify_lyrics(lyrics)
        genres.append(genre)
    return genres

# Încarcă setul de date de test
test_data = pd.read_csv('balanced_test_dataset.csv')

# Obține predicțiile pentru setul de test
test_lyrics = test_data['Cleaned_Lyrics']
predicted_genres = classify_lyrics_batch(test_lyrics)

# Genurile reale
true_genres = test_data['Genre']

# Calcularea metricei de evaluare
accuracy = accuracy_score(true_genres, predicted_genres)
precision, recall, f1, _ = precision_recall_fscore_support(true_genres, predicted_genres, average='weighted')

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")
```

```python
import pandas as pd
import joblib
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

# Încarcă modelul LDA și vectorizatorul
lda = joblib.load('lda_model.pkl')
vectorizer = joblib.load('vectorizer.pkl')

# Încarcă distribuția topicurilor pe genuri
topic_genre_distribution = pd.read_csv('topic_genre_distribution_for_predict.csv', index_col=0)

# Funcția pentru a prelucra și clasifica noile versuri
def classify_lyrics(lyrics):
    lyrics = " ".join(lyrics.split("\n")).strip()
    X_new = vectorizer.transform([lyrics])
    topic_distribution = lda.transform(X_new)
    dominant_topic = topic_distribution.argmax(axis=1)[0]
    genre = topic_genre_distribution.idxmax(axis=0)[dominant_topic]
    return genre

# Funcția pentru a clasifica un set de versuri
def classify_lyrics_batch(lyrics_list):
    genres = []
    for lyrics in lyrics_list:
        genre = classify_lyrics(lyrics)
        genres.append(genre)
    return genres

# Încarcă setul de date de test
test_data = pd.read_csv('balanced_test_dataset2.csv')

# Obține predicțiile pentru setul de test
test_lyrics = test_data['Cleaned_Lyrics']
predicted_genres = classify_lyrics_batch(test_lyrics)

# Genurile reale
true_genres = test_data['Genre']

# Calcularea metricei de evaluare
accuracy = accuracy_score(true_genres, predicted_genres)
precision, recall, f1, _ = precision_recall_fscore_support(true_genres, predicted_genres, average='weighted', zero_division=0)

print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")

# Distribuția genurilor în setul de test și în predicții
print("Distribuția genurilor în setul de test:")
print(true_genres.value_counts())
print("Distribuția genurilor prezise:")
print(pd.Series(predicted_genres).value_counts())
```
