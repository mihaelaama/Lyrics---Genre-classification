```python
import pandas as pd
import joblib
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Încarcă modelul LDA și vectorizatorul
lda = joblib.load('lda_model.pkl')
vectorizer = joblib.load('vectorizer.pkl')

# Încarcă distribuția topicurilor pe genuri
topic_genre_distribution = pd.read_csv('topic_genre_distribution_for_predict.csv', index_col=0)

# Funcția pentru a prelucra și clasifica noile versuri
def classify_lyrics(lyrics):
    lyrics = " ".join(lyrics.split("\n")).strip()
    X_new = vectorizer.transform([lyrics])
    topic_distribution = lda.transform(X_new)
    dominant_topic = topic_distribution.argmax(axis=1)[0]
    genre = topic_genre_distribution.idxmax(axis=0)[dominant_topic]
    return genre

# Funcția pentru a clasifica un set de versuri
def classify_lyrics_batch(lyrics_list):
    genres = []
    for lyrics in lyrics_list:
        genre = classify_lyrics(lyrics)
        genres.append(genre)
    return genres

# Încarcă setul de date de test
test_data = pd.read_csv('balanced_test_dataset.csv')

# Obține predicțiile pentru setul de test
test_lyrics = test_data['Cleaned_Lyrics']
predicted_genres = classify_lyrics_batch(test_lyrics)

# Genurile reale
true_genres = test_data['Genre']

# Calcularea matricei de confuzie
conf_matrix = confusion_matrix(true_genres, predicted_genres, labels=true_genres.unique())

# Crearea unui DataFrame pentru matricea de confuzie
conf_matrix_df = pd.DataFrame(conf_matrix, index=true_genres.unique(), columns=true_genres.unique())

# Salvarea matricei de confuzie într-un fișier CSV
conf_matrix_df.to_csv('confusion_matrix.csv')

# Salvarea matricei de confuzie într-un plot folosind seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=true_genres.unique(), yticklabels=true_genres.unique())
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.savefig('confusion_matrix.png')
plt.show()

# Print formatted output
print("\nConfusion Matrix:")
print(conf_matrix_df)
```
