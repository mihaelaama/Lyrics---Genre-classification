```python
import csv

# Function to print specified number of lines from the CSV file in columns
def print_lines_from_csv_in_columns(file_path, num_lines):
    with open(file_path, "r", encoding="utf-8") as csvfile:
        reader = csv.reader(csvfile)
        # Skip the header row if it exists
        next(reader, None)
        # Print the specified number of lines in columns
        for i, row in enumerate(reader):
            print("Artist:", row[0])
            print("Title:", row[1])
            print("Genre:", row[2])
            print("Cleaned_Lyrics:", row[3])
            print("=" * 50)  # Separator between records
            if i == num_lines - 1:
                break

# Specify the path to your CSV file
csv_file_path = "set_de_date_final.csv"

# Specify the number of lines you want to print
num_lines_to_print = 5

# Call the function to print the specified number of lines from the CSV file in columns
print_lines_from_csv_in_columns(csv_file_path, num_lines_to_print)

```

```python
import csv

# Function to count the occurrences of each genre from the CSV file
def count_genres_from_csv(file_path):
    genre_dict = {}
    with open(file_path, "r", encoding="utf-8") as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            genre = row['Genre']
            if genre not in genre_dict:
                genre_dict[genre] = 1
            else:
                genre_dict[genre] += 1
    return genre_dict

# Specify the path to your CSV file
csv_file_path = "set_de_date_final.csv"

# Call the function to count the occurrences of each genre
genre_counts = count_genres_from_csv(csv_file_path)

# Print the genre counts
for genre, count in genre_counts.items():
    print(f"{genre}: {count}")
```

```python
import csv
from nltk import word_tokenize

# Function to tokenize lyrics from the CSV file and write them to a text file
def tokenize_lyrics_from_csv(csv_file_path, output_file_path):
    with open(output_file_path, 'w') as g:
        with open(csv_file_path, "r", encoding="utf-8") as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                lyrics = row['Cleaned_Lyrics']
                if lyrics:
                    words = word_tokenize(lyrics)
                    outstring = ' '.join(words)
                    g.write(outstring)
                    g.write('\n\n')

# Specify the path to your CSV file and the path for the output text file
csv_file_path = "set_de_date_final.csv"
output_file_path = "alllyrics.txt"

# Call the function to tokenize the lyrics and write them to the output text file
tokenize_lyrics_from_csv(csv_file_path, output_file_path)
```

```python
import csv
import json

# Constants
min_number = 4000
genre_counts_file = 'genrecounts.txt'
filtered_genre_counts_file = 'top20filteredgenrecounts.txt'
input_data_csv = 'set_de_date_final.csv'  # Your input CSV file
output_filtered_json = 'top20indexed_data.json'

# Load and filter genre counts
genre_dict = {}
with open(genre_counts_file, 'r') as f:
    with open(filtered_genre_counts_file, 'w') as g:
        for line in f:
            split = line.strip().split()
            genre = ' '.join(split[:-1])
            count = int(split[-1])
            if count >= min_number:
                genre_dict[genre] = count
                g.write(f"{genre} {count}\n")

print("Finished filtering genres")

# Filter data points based on genre
with open(input_data_csv, 'r', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    with open(output_filtered_json, 'w') as g:
        for row in reader:
            genre = row['Genre']  # Adjusted to match the column name
            if genre in genre_dict:
                json.dump(row, g)
                g.write('\n')

print("Data points filtering completed")
```
